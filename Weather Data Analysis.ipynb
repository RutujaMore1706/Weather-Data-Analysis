## Visualization

### temperature vary over time 

```python
data['Date/Time'] = pd.to_datetime(data['Date/Time'])

# Sort the dataframe by the 'Date/Time' column
data.sort_values('Date/Time', inplace=True)

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(data['Date/Time'], data['Temp_C'])
plt.title('Temperature Variation Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.tight_layout()

plt
```

### distribution of dew point temperatures (Dew Point Temp_C) throughout the dataset

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Dew Point Temp_C'], bins=20, edgecolor='black')
plt.title('Distribution of Dew Point Temperatures')
plt.xlabel('Dew Point Temperature (°C)')
plt.ylabel('Frequency')
plt.tight_layout()

plt
```

### correlation between relative humidity (Rel Hum_%) and visibility (Visibility_km)

```python
plt.scatter(data['Rel Hum_%'], data['Visibility_km'])
plt.title('Correlation between Relative Humidity and Visibility')
plt.xlabel('Relative Humidity (%)')
plt.ylabel('Visibility (km)')
plt.tight_layout()


plt
```

### How does the wind speed (Wind Speed_km/h) change across different weather conditions

```python
plt.figure(figsize=(10, 15))
data.boxplot(column='Wind Speed_km/h', by='Weather Conditions', rot=90)
plt.title('Wind Speed Variation Across Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Wind Speed (km/h)')
plt.ylim(0, 50)


plt
```

### Atmospheric Pressure Over Time

```python
plt.plot(data['Date/Time'], data['Press_kPa'])
plt.title('Atmospheric Pressure Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Pressure (kPa)')
plt.xticks(rotation=90)


plt.show()
```

```python
weather_condition_counts = data['Weather Conditions'].value_counts()

# Create the bar plot for the most frequent weather conditions
plt.figure(figsize=(8, 6))
weather_condition_counts.plot(kind='bar')
plt.title('Most Frequent Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Frequency')
plt.tight_layout()

plt

# Create stacked bar plots to show the relationship between weather conditions and other variables
plt.figure(figsize=(12, 6))
data.groupby('Weather Conditions')['Temp_C'].mean().plot(kind='bar', color='blue', label='Temperature (°C)')
data.groupby('Weather Conditions')['Rel Hum_%'].mean().plot(kind='bar', color='orange', label='Relative Humidity (%)')
data.groupby('Weather Conditions')['Wind Speed_km/h'].mean().plot(kind='bar', color='green', label='Wind Speed (km/h)')
plt.title('Weather Conditions and Related Variables')
plt.xlabel('Weather Conditions')
plt.ylabel('Mean Value')
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()

# Display the plot
plt
```
